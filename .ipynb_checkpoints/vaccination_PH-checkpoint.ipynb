```python
import plotly.express as px
import pandas as pd

df = pd.read_csv('country_vaccinations.csv')
```

```python
dfFilter = df[df.country == 'Philippines']
```

```python
fig = px.line(dfFilter, x='date', y="total_vaccinations")
fig.show()
```

```python
fig = px.line(dfFilter, x='date', y="daily_vaccinations")
fig.show()
```
